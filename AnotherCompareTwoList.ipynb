```python
from difflib import unified_diff

def compare_texts(text1, text2):

    lines1 = text1.splitlines()
    lines2 = text2.splitlines()

    diff = unified_diff(
        lines1,
        lines2,
        fromfile='Text_A',
        tofile='Text_B',
        lineterm=''
    )

    return "\n".join(diff)

text_a = """
"""
text_b = """
"""

print(compare_texts(text_a, text_b))
```
