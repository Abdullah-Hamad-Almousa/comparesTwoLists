```python
def parse_libs(text):

    libs = {}
    for line in text.splitlines():
        line = line.strip()
        if "=" in line:
            name, version = line.split("=", 1)
            libs[name.strip()] = version.strip().strip('"')
    return libs


def compare_libs(my_libs_text, other_libs_text):
    my_libs = parse_libs(my_libs_text)
    other_libs = parse_libs(other_libs_text)

    missing_in_me = []
    missing_in_other = []
    version_diff = []

    for lib in other_libs:
        if lib not in my_libs:
            missing_in_me.append(lib)
        else:
            if my_libs[lib] != other_libs[lib]:
                version_diff.append((lib, my_libs[lib], other_libs[lib]))

    for lib in my_libs:
        if lib not in other_libs:
            missing_in_other.append(lib)

    return missing_in_me, missing_in_other, version_diff


# ---------------- TEST EXAMPLE ----------------
myList = """

"""

compareList = """

"""

missing_in_me, missing_in_other, version_diff = compare_libs(myList, compareList)

print("Missing in me:", missing_in_me)
print("Missing in other person:", missing_in_other)
print("Version differences:", version_diff)
```
